```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] =     8
plt.rcParams["font.family"] = "Times New Roman"

# load configuration table
RESULTS_FILE = pd.read_excel("Results.xlsx", "Sheet1")
RES_TAB = RESULTS_FILE.rename(columns={RESULTS_FILE.columns[0]:"# CONFIG"})    
RES_TAB
```

```python
SUB_RES_TAB = RES_TAB[RES_TAB['# CONFIG'].isin([0,*np.arange(1001,1326)])]
SUB_RES_TAB
```

```python
polar_data_idx = SUB_RES_TAB.columns[[0,22,4,10,1,19]]
# polar_data_idx = SUB_RES_TAB.columns[[0,22,4,10,1]]
polar_data=SUB_RES_TAB[polar_data_idx]
polar_data
```

```python
df = pd.DataFrame(polar_data)
for i in df.columns:
    df[i] = abs(df[i]-df[i][0])
```

```python
df_new = df[df['# CONFIG'].isin([*np.arange(1001,1326)])]
df_new
```

```python
for i in df_new.columns[1:]:
    # print(max(df_new[i]))
    df_new[i] = 1-((df_new[i]-min(df_new[i]))/max(df_new[i]-min(df_new[i])))
    # df_new[i] = (df_new[i])/max(df_new[i])

```

```python
df_new
```

```python
lista = df_new.iloc[:,1:].product(axis='columns')
print(lista)
```

```python
df_new
```

```python
df_new.insert(1,"vote",lista)
```

```python
df_new
```

```python
df_new.sort_values(by=['vote'], inplace=True, ascending=False)
df_new
```

```python
categories = ["Root Mean Square Error of Inter-Spike Interval","Mean Firing Rate","Mean Bursting Rate","Pearson's Coefficient","Burstiness Index"]

both=[]
# lin_col = []
for i in range(len(df_new)):
    both.append(pd.DataFrame(dict(
        r=df_new.iloc[i,2:].to_numpy(),
        theta=categories,
        Network='All')))
    # lin_col.append('')

choosen_config = [1043, 1155, 1196]
plot_data = df_new[df_new['# CONFIG'].isin(choosen_config)]
plot_data
```

```python

for i in range(len(choosen_config)):
    both.append(pd.DataFrame(dict(
        r=plot_data[plot_data['# CONFIG'].isin([choosen_config[i]])].iloc[:,2:].to_numpy()[0],
        theta=categories,
        Network='SNN'+str(i+1))))
both
```

```python
df_both = pd.concat(both, axis=0)
df_both
```

```python
import plotly.express as px
fig = px.line_polar(df_both, r='r', color='Network', theta='theta', line_close=True,
                    color_discrete_sequence=['rgba(180,180,180,0.5)','rgb(255,1,1)','rgb(150,1,100)','rgb(1,1,255)'], width = int(950), height = int(700))
fig.update_layout(
    template=None,
    font=dict(
        family="Helvetica",
        size=22
    )
)
fig.update_traces(line={'width': 5},  )
fig.show()
```

```python
import plotly.io as pio
#save a figure of 300dpi, with 1.5 inches, and  height 0.75inches
pio.write_image(fig, "./paperimages/test.svg", width=950, height=700, scale=1)
pio.write_image(fig, "./paperimages/test.png", width=950, height=700, scale=1)
```
